```python
import requests
import http.client
import pandas as pd
import json
df = pd.read_excel("api_test_data.xlsx")
print(df)
```

```python
URL = "127.0.0.1:8000"
connection = http.client.HTTPConnection(URL)
headers = {"Content-type": "application/json", "Accept": "application/json"}
params = json.dumps({"email": "user@example.com", "password":"testpass123"})
request = connection.request("POST","/api/user/token/", params, headers)


response = connection.getresponse()
data = json.loads(response.read().decode())
connection.close()

token = data['token']
print(token)
```

```python
connection = http.client.HTTPConnection(URL)
headers = {"Content-type": "application/json", "Accept": "application/json", "Authorization": f"token {token}"}
request = connection.request("GET","/api/horse/horses/", "", headers)
response = connection.getresponse()
data = json.loads(response.read().decode())
api_key = data[0]['api_key']
connection.close()
print(api_key)
```

```python
for i, dp in enumerate(df.iterrows()):
    connection = http.client.HTTPConnection(URL)
    data_point = df[['api_key', 'date_created', 'gps_lat', 'gps_long', 'temp', 'hr', 'hr_interval']].iloc[i].to_dict()
    data_point['api_key'] = api_key
    params = json.dumps(data_point)
    headers = {"Content-type": "application/json", "Accept": "application/json", "Authorization": f"token {token}"}
    request = connection.request("POST","/api/horse/datapoints/", params, headers)
    response = connection.getresponse()
    data = json.loads(response.read().decode())
    connection.close()
    print(data)
```
